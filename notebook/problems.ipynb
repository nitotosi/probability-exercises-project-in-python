### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import numpy as np

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'  
  
        
#Initialize important variables
n_simulations = 10000
threeWhiteTwoRedFrequency = 0
allFiveSameColorFrequency = 0

for i in range(n_simulations):
    #make a list of the colors that we choose
    chosenColors = [d[np.random.randint(60)],d[np.random.randint(60)],d[np.random.randint(60)],d[np.random.randint(60)],d[np.random.randint(60)]]
    

    #Convert it to a numpy array
    #chosenColors = np.array(chosenColors)
    #print(chosenColors)

    #find the number of each taht we picked
    whites = chosenColors.count('White')
    reds = chosenColors.count('Red')
    greens = chosenColors.count('Green') 
    
    #Keep track if the combination met the criteria
    if whites == 3 or reds == 2:
        threeWhiteTwoRedFrequency += 1
    
    if all(i == chosenColors[0] for i in chosenColors):
        allFiveSameColorFrequency += 1
    
    
#print 'the probability of 3 white and 2 red is' in percentage
print(threeWhiteTwoRedFrequency/n_simulations)

#print 'the probability of all the same color is' in percentage   
print(allFiveSameColorFrequency/n_simulations)
```
